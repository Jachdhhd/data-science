```python
import csv
from copy import deepcopy as dcopy


funnel_template = {'1_home_page':0, '2_search_page':0, '3_payment_page':0, '4_payment_confirmation_page':0}
funnel_template_date = {'2015-01': funnel_template.copy(),'2015-02': funnel_template.copy(), 
'2015-03':funnel_template.copy(), '2015-04': funnel_template.copy()}
funnel_by_month = {} # создаем словарь, где будут храниться словари за каждый месяц
funnel_by_device = {}
funnel_by_gender = {}
funnel_template_device = {'Mobile':dcopy(funnel_template_date),'Desktop':dcopy(funnel_template_date)}
with open('click_stream3.csv', mode='r') as csv_file: # открываем файл
    csv_reader = csv.DictReader(csv_file, fieldnames=['ID','page','date']) # читаем файл
    
    # проходим циклом по объекту csv_reader
    # берем строку из объекта csv_reader
    
    for row in csv_reader:
        page = list(row.items())[1][1] # вытаскиваем название страницы
        event_date = list(row.items())[2][1][:-3] # вытаскиваем дату в формате YYYY-MM
        device = list(row.items())[3][1][0]
        gender = list(row.items())[3][1][1]
                # проверяем существует ли ключ со значением event_date в словаре
        if event_date not in funnel_by_month: # если нет
            funnel_by_month[event_date] = funnel_template.copy() # создаем ключ, в значение копируем заготовку
        if device not in funnel_by_device : # если нет
            funnel_by_device[device] = dcopy(funnel_template_date)
        if gender not in funnel_by_gender : # если нет
            funnel_by_gender[gender] = dcopy(funnel_template_device)
            
            
        if page == '1_home_page':
            funnel_by_month[event_date]['1_home_page'] += 1
        elif page == '2_search_page':
            funnel_by_month[event_date]['2_search_page'] += 1
        elif page == '3_payment_page':
            funnel_by_month[event_date]['3_payment_page'] += 1
        else:
            funnel_by_month[event_date]['4_payment_confirmation_page'] += 1 
        
             
            
        if page == '1_home_page':
            funnel_by_device[device][event_date]['1_home_page'] += 1
        elif page == '2_search_page':
            funnel_by_device[device][event_date]['2_search_page'] += 1
        elif page == '3_payment_page':
            funnel_by_device[device][event_date]['3_payment_page'] += 1
        else:
            funnel_by_device[device][event_date]['4_payment_confirmation_page'] += 1
            
            
        if page == '1_home_page':
            funnel_by_gender[gender][device][event_date]['1_home_page'] += 1
        elif page == '2_search_page':
            funnel_by_gender[gender][device][event_date]['2_search_page'] += 1
        elif page == '3_payment_page':
            funnel_by_gender[gender][device][event_date]['3_payment_page'] += 1
        else:
            funnel_by_gender[gender][device][event_date]['4_payment_confirmation_page'] += 1

funnel_by_gender 
# Несмотря на то, что женщины в январе делали большее количество покупок независимо от устройства, 
#однако с февраля кол-во женщин и мужчин, подтвердивших покупку через мобильное устройство сравнялось, подвердивших покупку 
#через десктоп мужчин с февраля по апрель больше, чем женщин. К тому же с января по апрель количество женщин, совершивших 
#покупку через мобильное устройство сократилось почти в 7 раз, тогда как кол-во мужчин в 5 раз. Исходя из этого можно сделать 
# вывод о том, что изменения на сайте повлияли в большей степени на женщин.

            
```
