```python
ds_homework = df_merged2.copy()
ds_homework1 = ds_homework[ds_homework['is_homework'] == True] 
student_co = ds_homework[ds_homework['is_homework'] == True] 
student_co['start_date'] = pd.to_datetime(student_co['start_date'],utc = True).dt.floor('D')
student_co['start_date'] = student_co['start_date'].values.astype('datetime64[M]')
student_co1 = student_co.groupby(['student_id','title']).agg({'start_date':'min'}).reset_index()
student_co2 = student_co1.groupby(['title','start_date']).agg({'student_id':'count'})
student_co3 = student_co2.unstack().T.fillna(value =0).T.stack().reset_index()
grouped4 = student_co3.groupby('title')

fig, axes = plt.subplots(15,1, figsize=(10,70))  #рисуем 15 графиков прироста студентов для каждого курса
for (name, dtf), ax in zip(grouped4 , axes.flat):
    
    dtf.plot(x='start_date',y='student_id', title = name, rot = 90, color = 'b', ax=ax)
    ax.yaxis.label.set_size(15)
    ax.xaxis.label.set_size(15)
    ax.title.set_size(15)
fig.tight_layout()
plt.show()
```

```python

ds_progress1 = ds_homework1.copy()

ds_progress1['start_date'] = ds_progress1['start_date'].values.astype('datetime64[M]')
ds_progress1['finish_date'] = pd.to_datetime(ds_progress1['finish_date'],utc = True).dt.floor('D')
ds_progress1['finish_date'] = ds_progress1['finish_date'].values.astype('datetime64[M]')
ds_progress1['finish_date'] = ds_progress1['finish_date'].fillna(value=0)



for dt in pd.date_range("2016-01-01", "2019-07-30", freq="MS"):
    new_column = str.format("{}-{:02d}", dt.year, dt.month)
    ds_progress1[new_column] = 0

for ind, row in ds_progress1.iterrows():
    start  = row['start_date']                               #находим количество прогрессов по выполнению домашних работ в каждом месяце за всю историю
    finish = row['finish_date']
    dtRange = pd.date_range(start, finish, freq="MS")
    for dt in dtRange:
        colName = str.format("{}-{:02d}", dt.year, dt.month)
        ds_progress1.loc[ind, colName] = 1

ds_progress3 = ds_progress1.groupby('title').sum()
ds_progress4 = ds_progress3.loc[:, '2016-01':]

```
