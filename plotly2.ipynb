```python
import plotly.express as px
import pandas_datareader as pdr 
import pandas as pd
```

```python
data = pdr.get_data_yahoo(symbols = 'TSLA')
data = data.reset_index()
df = data.loc[data['Date'].dt.year==2020]
```

```python
fig = px.line(df, x='Date', y = ['Low','High'])
fig.show()
```

```python
   350  
```

```python
data.head()
```

```python
data['diff'] = (data['Close']-data['Open'])/data['Close']
data.head()
```

```python
fig = px.line(data, x='Date', y = ['diff'])
fig.show()
```

```python
   5  2020 - 12%.    3  2020 - 13.6%
```

```python
df1 = data.groupby(pd.Grouper(key="Date", freq="M")).mean()
```

```python
fig = px.line(df1, x=df1.index, y = 'Volume', hover_data=['Close'])
fig.show()
```
