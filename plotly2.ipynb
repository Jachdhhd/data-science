```python
import plotly.express as px
import pandas_datareader as pdr 
import pandas as pd
```

```python
data = pdr.get_data_yahoo(symbols = 'TSLA')
data = data.reset_index()
df = data.loc[data['Date'].dt.year==2020]
```

```python
fig = px.line(df, x='Date', y = ['Low','High'])
fig.show()
```

```python
Впервые цена превысила 350 в Июле
```

```python
data.head()
```

```python
data['diff'] = (data['Close']-data['Open'])/data['Close']
data.head()
```

```python
fig = px.line(data, x='Date', y = ['diff'])
fig.show()
```

```python
Самое резкое понижение 5 Февраля 2020г - 12%. Самое резкое повышение 3 февраля 2020г - 13.6%
```
